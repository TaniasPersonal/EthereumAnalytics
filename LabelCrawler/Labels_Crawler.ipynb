```python
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
import csv 
```

```python
def writefile(filename, labels_list):
# opening the csv file in 'w' mode 
    file = open(filename, 'w', newline ='') 
    with file: 
        # identifying header   
        header = ['Address', 'Entity', 'Balance', "Total_tx"] 
        writer = csv.DictWriter(file, fieldnames = header)   
        # writing data row-wise into the csv file 
        writer.writeheader() 
        for row in labels_list:
            writer.writerow({'Address' : row[0],  
                             'Entity': row[1],  
                             'Balance': row[2], 
                             'Total_tx': row[3]}) 
        
```

```python
def extracturls(links):
    link_urls = [link.get_attribute('href') for link in links]
#     print(link_urls)
    labels_urls = []
    for c in link_urls:
        if 'https://etherscan.io/accounts/label/' in c:
            labels_urls.append(c)
    print(labels_urls)
    return labels_urls
```

```python
for index in range (19,len(labels_urls)):
    url = labels_urls[index]
    print(url)
    browser.get(url)
    print(url)
    df= []
    try:
        table_id = browser.find_element(By.ID, 'table-subcatid-1')
        ID = "table-subcatid-1"
    except:
        table_id = browser.find_element(By.ID, 'table-subcatid-0')
        ID = "table-subcatid-0"
    rows = table_id.find_elements(By.TAG_NAME, "tr") # get all of the rows in the table
    for row in rows:
        # Get the columns (all the column 2)        
        cols = row.find_elements(By.TAG_NAME, "td") #note: index start from 0, 1 is col 2
        one_col = []
        for col in cols:
            one_col.append(col.text)
        df.append(one_col)
    pages = browser.find_element(By.XPATH, '//*[@id="'+ID+'_paginate"]/ul/li[3]/span/strong[2]').text
    pages = int(pages)
    count = 25;
    for i in range(1,pages):
        browser.get(url+"/"+str(i)+"?"+"subcatid=undefined&size=25&start="+str(count)+"&col=1&order=asc")
        count +=25
        table_id = browser.find_element(By.ID, ID)
        rows = table_id.find_elements(By.TAG_NAME, "tr") # get all of the rows in the table
        for row in rows:
            # Get the columns (all the column 2)        
            cols = row.find_elements(By.TAG_NAME, "td") #note: index start from 0, 1 is col 2
            one_col = []
            for col in cols:
                one_col.append(col.text)
            df.append(one_col)

    print(len(df))
    cat_labels  = [x for x in df if x != []]
    print(cat_labels)
    writefile("labelled/"+url.split("/")[-1]+".csv",cat_labels) #Create a folder named labelled in same folder
```
